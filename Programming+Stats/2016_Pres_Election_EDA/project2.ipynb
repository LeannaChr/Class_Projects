```python
candidates = df['candidates'].value_counts()
candidates
```

```python
DT = df[df['candidates'] == 'Donald Trump']
HC = df[df['candidates'] == 'Hillary Clinton']
BS = df[df['candidates'] == 'Bernie Sanders']

DT_pro = DT[DT['message'] == 'pro']
DT_con = DT[DT['message'] == 'con']

HC_pro = HC[HC['message'] == 'pro']
HC_con = HC[HC['message'] == 'con']

BS_pro = BS[BS['message'] == 'pro']
BS_con = BS[BS['message'] == 'con']
```

```python
DT.shape
```

```python
DT_counts = DT['location'].value_counts()
HC_counts = HC['location'].value_counts()
BS_counts = BS['location'].value_counts()

DT_p_counts = DT_pro['location'].value_counts()
DT_c_counts = DT_con['location'].value_counts()

HC_p_counts = HC_pro['location'].value_counts()
HC_c_counts = HC_con['location'].value_counts()

BS_p_counts = BS_pro['location'].value_counts()
BS_c_counts = BS_con['location'].value_counts()
```

```python
X = ['Donald Trump','Hillary Clinton','Bernie Sanders']
pro = [13368,30598,11353]
con = [40493,13778,263]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, pro, 0.4, label = 'Pro')
plt.bar(X_axis + 0.2, con, 0.4, label = 'Con')
  
plt.xticks(X_axis, X)
plt.xlabel("Popular Candidates")
plt.ylabel("Number of Pro and Con Per Candidate")
plt.title("Number of Ads Aired")
plt.legend()
plt.show()
```
